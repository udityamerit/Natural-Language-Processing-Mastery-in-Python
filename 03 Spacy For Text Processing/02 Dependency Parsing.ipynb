# Dependency Parsing

```python
import spacy
from prettytable import PrettyTable
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
text = "Spot Intelligence is a NLP company that builds custom models"
```

```python
doc = nlp(text)
doc
```

```python
table = PrettyTable()
table.field_names = ['token', 'dep', 'head text', 'head pos', 'children']

for token in doc:
        children = [child.text for child in token.children]
        table.add_row([token.text, token.dep_, token.head.text, token.head.pos_, children])

print(table)
```

```python
from spacy import displacy
displacy.render(doc, style = 'dep', jupyter=True, options={})
```

```python
from spacy import displacy
displacy.render(doc, style = 'dep', jupyter=True, options={'distance':85, 'compact':True, 'bg':'#09a3d3', 'color': '#ffffff'})
```
