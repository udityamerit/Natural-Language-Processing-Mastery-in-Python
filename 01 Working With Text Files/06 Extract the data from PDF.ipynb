## Reading all the pages

```python
pages = []
with open('data/ML.pdf', 'rb') as fp:
        reader = PyPDF2.PdfReader(fp)
        for i in range(len(reader.pages)):
                page = reader.pages[i]
                text = page.extract_text()
                pages.append(text)
```

```python
pages
```

```python
text = '\n'.join(pages)
with open('data/ml.txt', 'w') as fp:
        fp.write(text)
```

## reading only selected pages

```python
pages = []
with open('data/ML.pdf', 'rb') as fp:
        reader = PyPDF2.PdfReader(fp)
        for i in range(len(reader.pages)):
                if i+1 in [1,2]:
                        page = reader.pages[i]
                        text = page.extract_text()
                        pages.append(text)

text = '\n'.join(pages)
with open('data/ml1.txt', 'w') as fp:
        fp.write(text)
```
