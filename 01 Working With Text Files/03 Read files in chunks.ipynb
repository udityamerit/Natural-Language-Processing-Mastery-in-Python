```python
url = 'https://github.com/laxmimerit/All-CSV-ML-DATA-Files-Download/raw/master/IMDB-Dataset.csv'
df = pd.read_csv(url, chunksize=10000)
data = pd.DataFrame()
for chunk in df:
        data = pd.concat([data, chunk])
```
