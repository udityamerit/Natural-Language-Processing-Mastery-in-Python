```python
recognizer = sr.Recognizer()
with sr.Microphone() as source:
        print('Say Something...')
        audio1 = recognizer.listen(source)
```

```python
text = recognizer.recognize_google(audio1)
```

```python
print(text)
```

```python
with open('audio/record2.wav', 'wb') as fp:
        fp.write(audio1.get_wav_data())
```

## Recording for some time limit

```python
with sr.Microphone() as source:
        print('Recoding for 10 sec...')
        audio2 = recognizer.record(source, duration=10)
        text = recognizer.recognize_google(audio2)

```

```python
print(text)
```

```python
with open('audio/record3.wav', 'wb') as fp:
        fp.write(audio2.get_wav_data())
```

## Read the audio and convert in text data

```python
with sr.AudioFile('audio/record1.wav') as fp:
        audio_data = recognizer.record(fp)
        text = recognizer.recognize_google(audio_data)
        
print(text)
```

## Read the audio for specific time duration

```python
with sr.AudioFile('audio/record1.wav') as fp:
        audio_data = recognizer.record(fp, duration=3)
        text = recognizer.recognize_google(audio_data)
        
print(text)
```
