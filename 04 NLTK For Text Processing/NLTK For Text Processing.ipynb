# Stemming

```python
from nltk.stem import PorterStemmer
text = 'The scientists discover new species every year, Last year, they discovered an ancient artifact. they are discovering new techniques with their recent discovery'
```

```python
stemmer = PorterStemmer()
words = word_tokenize(text)
stemmed_words = [stemmer.stem(word) for word in words]

print(stemmed_words)
```

```python
from prettytable import PrettyTable
table = PrettyTable()
table.field_names = ['word', 'stemmed word']
for word, stw in zip(words, stemmed_words):
        table.add_row([word, stw])

print(table)
```

# Lemmatizing

```python
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

```python
lemmetizer = WordNetLemmatizer()
```

```python
words = word_tokenize(text)
```

```python
lemma = [lemmetizer.lemmatize(word) for word in words]
```

```python
table = PrettyTable(field_names=['word', 'stemmed word', 'lemma word'])

for word, stm, lm in zip(words, stemmed_words, lemma):
        table.add_row([word, stm, lm])

print(table)
```

# Named Entity Recognition(NER)

```python
from nltk.chunk import ne_chunk
```

```python
nltk.download('words')
nltk.download('maxent_ne_chunker_tab')
```

```python
text = """
On December 20, 2025, Satya Nadella announced that Microsoft would invest $5 billion to build a new research hub in London. This facility, located near Kings Cross, aims to rival the innovation seen at Google in Mountain View, California.
"""

words = word_tokenize(text)

tagged_words = pos_tag(words)
```

```python
ner_tree = ne_chunk(tagged_words)
```

```python
ner_tree
```

```python
for ner in ner_tree:
        if hasattr(ner, 'label'):
                print(' '.join(c[0] for c in ner), ":", ner.label())
```
